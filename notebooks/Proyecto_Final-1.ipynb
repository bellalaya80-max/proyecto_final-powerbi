## Configuración del Entorno y Librerías
Comenzamos importando las herramientas necesarias para el análisis y limpieza para posteriormente exportar y realizar la visualización en Power BI.

```python
import pandas as pd
import numpy as np
```

```python
# Cargar los datasets 
df_salud = pd.read_csv(r"H:\\Mi unidad\\BELLA\\CURSO DATA ANALYTICS\\Proyecto_Final\\data\\brfss2022.csv")
df_zillow = pd.read_csv(r"H:\\Mi unidad\\BELLA\\CURSO DATA ANALYTICS\\Proyecto_Final\\data\\ZHVI.csv") 

# Exploración del BRFSS (Buscamos la columna del estado, suele llamarse '_STATE' o 'STATE')
print("DATOS DE SALUD (BRFSS)")
print(f"Filas iniciales: {df_salud.shape[0]}") # Número de filas (registros) en el dataset de salud
print("Primeras 20 columnas:")
print(list(df_salud.columns[:20]))

# Exploración de ZILLOW (Buscamos la columna del estado, suele llamarse 'RegionName' o 'StateName')
print("\n DATOS DE VIVIENDA (ZILLOW)")
print(f"Filas iniciales: {df_zillow.shape[0]}")
print("Primeras 15 columnas:")
print(list(df_zillow.columns[:15]))

print("Así se ve el archivo por dentro:")
display(df_zillow.head())

print("Primeras 5 encuestas (filas):")
display(df_salud.head())
```

```python
# LIMPIEZA DE ZILLOW 
# Renombramos esa columna sin nombre a 'Fecha'
df_zillow = df_zillow.rename(columns={'Unnamed: 0': 'Fecha'})

# Derretimos para tener: Fecha | Estado | Precio_Vivienda
df_zillow_long = df_zillow.melt(id_vars=['Fecha'], var_name='StateName', value_name='Precio_Vivienda')

# Calculamos el precio promedio histórico por estado 
df_precios_estado = df_zillow_long.groupby('StateName')['Precio_Vivienda'].mean().reset_index()


# EL TRADUCTOR DE ESTADOS (Diccionario FIPS)
# Este diccionario convierte los números del gobierno a nombres que Zillow entiende
fips_dict = {
    1: 'Alabama', 2: 'Alaska', 4: 'Arizona', 5: 'Arkansas', 6: 'California', 8: 'Colorado', 
    9: 'Connecticut', 10: 'Delaware', 11: 'District of Columbia', 12: 'Florida', 13: 'Georgia', 
    15: 'Hawaii', 16: 'Idaho', 17: 'Illinois', 18: 'Indiana', 19: 'Iowa', 20: 'Kansas', 
    21: 'Kentucky', 22: 'Louisiana', 23: 'Maine', 24: 'Maryland', 25: 'Massachusetts', 
    26: 'Michigan', 27: 'Minnesota', 28: 'Mississippi', 29: 'Missouri', 30: 'Montana', 
    31: 'Nebraska', 32: 'Nevada', 33: 'New Hampshire', 34: 'New Jersey', 35: 'New Mexico', 
    36: 'New York', 37: 'North Carolina', 38: 'North Dakota', 39: 'Ohio', 40: 'Oklahoma', 
    41: 'Oregon', 42: 'Pennsylvania', 44: 'Rhode Island', 45: 'South Carolina', 46: 'South Dakota', 
    47: 'Tennessee', 48: 'Texas', 49: 'Utah', 50: 'Vermont', 51: 'Virginia', 53: 'Washington', 
    54: 'West Virginia', 55: 'Wisconsin', 56: 'Wyoming'
}

# Creamos una nueva columna en salud con el nombre real
df_salud['StateName'] = df_salud['_STATE'].map(fips_dict)

# LA GRAN UNIÓN
# Cruzamos las 445k filas de salud con los 50 precios de Zillow
df_final = pd.merge(df_salud, df_precios_estado, on='StateName', how='inner')

print(f"✅ ¡Fusión exitosa! Tienes {df_final.shape[0]} filas listas para analizar.")
print(f"Y tienes {df_final.shape[1]} columnas.")
```

```python
# Definimos las columnas "estrella" para el análisis 

columnas_finales = [
    'StateName',        # El estado (Tu ancla geográfica)
    'Precio_Vivienda',  # Costo de vida (Tu dato estrella de Zillow)
    'MENTHLTH',         # Días con mala salud mental en el mes (Tu métrica de "Felicidad/Estrés")
    'PHYSHLTH',         # Días con mala salud física
    'EMPLOY1',          # Situación laboral (Empleado, desempleado, tiempo parcial...)
    'INCOME3',          # Nivel de ingresos familiares
    'MARITAL',          # Estado civil
    'EDUCA',            # Nivel educativo
    '_AGEG5YR',         # Grupo de edad
    'SEXVAR',           # Género
    'SLEPTIM1',         # Horas de sueño al día
    'EXERANY2',         # ¿Hizo ejercicio este mes?
    'HLTHPLN1',         # ¿Tiene seguro médico?
    '_BMI5CAT',         # Categoría de Índice de Masa Corporal
    'SMOKE100',         # Historial de tabaquismo (Factor de estrés/salud)
    'DRNKANY6',         # Consumo de alcohol reciente
    '_RACE',            # Raza / Etnia
    'VETERAN3',         # Estatus de Veterano
    'DIFFWALK',         # Dificultad para caminar (Indicador de discapacidad)
    'DIABETE4',         # Diagnóstico de diabetes (Enfermedad crónica)
    'ASTHMA3',          # Diagnóstico de asma
    'RENTHOM1'          # ¿Alquila o es dueño de su casa? 
    'IMONTH',           # Mes en que se hizo la encuesta
    'IYEAR'             # Año de la encuesta
]

# Filtrado Seguro
# Usamos una lista de comprensión para evitar errores si algún código cambió levemente
columnas_existentes = [col for col in columnas_finales if col in df_final.columns]

# Creamos nuestro DataFrame definitivo blindado con .copy()
df_powerbi = df_final[columnas_existentes].copy()

# Verificamos la rúbrica por última vez
print("DATASET DEFINITIVO PARA POWER BI")
print(f"✅ FILAS TOTALES: {df_powerbi.shape[0]}")
print(f"✅ COLUMNAS TOTALES: {df_powerbi.shape[1]}")

# EL GRAN EXPORT (Guardamos el archivo final)
df_powerbi.to_csv('dataset_powerbi_salud_vivienda.csv', index=False)
print("\n🎉 ¡Archivo 'dataset_powerbi_salud_vivienda.csv' guardado y listo!")
```
